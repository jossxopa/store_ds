```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

```

```python
from matplotlib import pyplot as plt
```

```python
from matplotlib.ticker import StrMethodFormatter
from matplotlib.lines import lineStyles
from IPython.core.display import Markdown
```

# **1. Análisis de facturación**

```python
total_ganancia_por_tienda = {
    "tienda 1": tienda["Precio"].sum(),
    "tienda 2": tienda2["Precio"].sum(),
    "tienda 3": tienda3["Precio"].sum(),
    "tienda 4": tienda4["Precio"].sum()
}

for numero_de_tienda, total in total_ganancia_por_tienda.items(): #.item dice: oye no solo me des la llave pasame todo lo que hay ahí, o sea la operación matematica
  print(f'El ingreso total de la {numero_de_tienda} es de: ${total:,.2f}') #,.2f le ponemos comas a los miles y solo 2 decimales

# Graficamos

nombres_tiendas = list(total_ganancia_por_tienda.keys())
ingresos_tiendas = list(total_ganancia_por_tienda.values())
plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas,ingresos_tiendas, color="slategray")

plt.title("Ingresos por tienda", fontsize=14)
plt.xlabel("Tiendas")
plt.ylabel("Ingresos (COP)")
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
plt.show()
```

# **2. Ventas por categoría**







     




```python
# tienda.isnull().sum() comprobamos que no hay datos nulos.

cantidad_productos_vendidos = {
   "tienda 1": tienda['Categoría del Producto'].value_counts().head(3).reset_index(),
   "tienda 2":tienda2['Categoría del Producto'].value_counts().head(3).reset_index(),
   "tienda 3":tienda3['Categoría del Producto'].value_counts().head(3).reset_index(),
   "tienda 4":tienda4['Categoría del Producto'].value_counts().head(3).reset_index()
}

colores_dona = ['#0A2E50', '#4A7C9D', '#D4C1AD'] #Para la gráfica

for nombre_tienda, cantidad_productos in cantidad_productos_vendidos.items():
  print(f'\n---Las 3 categorías más vendidas en la {nombre_tienda} son:---')
  print(cantidad_productos)
#Grafico dentro del for para que salga la gráfica.
  plt.figure(figsize=(4.5, 4.5))
  patches, texts, autotexts = plt.pie(cantidad_productos.iloc[:,1],
                                      labels=cantidad_productos.iloc[:,0],
                                      colors=colores_dona,
                                      autopct='%1.1f%%',
                                      startangle=90,
                                      pctdistance=0.85,
                                      wedgeprops={'edgecolor': 'white', 'linewidth': 1.5})
  # Circulo de la dona
  centro_circulo = plt.Circle((0, 0), 0.70, fc='white')
  fig = plt.gcf()
  fig.gca().add_artist(centro_circulo)
  plt.title(f"Top 3 Categorías en {nombre_tienda.title()}", fontsize=12, fontweight="bold", pad=20)
  plt.axis('equal')
  plt.tight_layout()
  #Cambiar el color de los números para que resalte
  for i, a in enumerate(autotexts):
    if i < 2:
      a.set_color('white')
      a.set_weight('bold')
    else:
      a.set_color('#333333')
      a.set_weight('bold')
  plt.show()

```

#**3. Calificación promedio de la tienda**

```python
promedio_por_tienda = {
    "tienda 1": tienda["Calificación"].mean().round(2),
    "tienda 2": tienda2["Calificación"].mean().round(2),
    "tienda 3": tienda3["Calificación"].mean().round(2),
    "tienda 4": tienda4["Calificación"].mean().round(2)
}
# Ordenar el diccionario para que la respuesta se ordene de mayor a menor

promedio_ordenado = sorted(promedio_por_tienda.items(), key=lambda item: item[1], reverse=True)

for nombre_de_tienda, promedio_de_tienda in promedio_ordenado:
  print(f'La satisfacción del cliente en relación con los productos vendidos de la {nombre_de_tienda} es de {promedio_de_tienda} ')

#Gráfico

nombres = [item[0] for item in promedio_ordenado]
valor = [item[1] for item in promedio_ordenado]

plt.bar(nombres, valor, color="skyblue")
plt.title("Promedio de calificación por tienda")
plt.ylabel("Promedio")
plt.xlabel("Tiendas")
plt.show()
```

#**4. Productos más y menos vendidos**

```python
maximo_y_minimo = {
    "tienda 1": pd.concat([tienda['Producto'].value_counts().head(3), tienda['Producto'].value_counts().tail(3)]).reset_index(),
    "tienda 2": pd.concat([tienda2['Producto'].value_counts().head(3), tienda2['Producto'].value_counts().tail(3)]).reset_index(),
    "tienda 3": pd.concat([tienda3['Producto'].value_counts().head(3), tienda3['Producto'].value_counts().tail(3)]).reset_index(),
    "tienda 4": pd.concat([tienda4['Producto'].value_counts().head(3), tienda4['Producto'].value_counts().tail(3)]).reset_index()
}

color_grafico = ['#0A2E50', '#0A2E50', '#0A2E50', '#95A5A6', '#95A5A6', '#95A5A6']

for titulo_tienda, tabla in maximo_y_minimo.items():
  print(f'\n --- Productos más y menos vendidos de la {titulo_tienda} son:  ')
  print(f'{tabla}')

  #Gráfico
  plt.figure(figsize=(12, 6))
  barras = plt.barh(tabla.iloc[:, 0], tabla.iloc[:, 1], color=color_grafico)
  plt.title(f"Productos más y menos vendidos de la {titulo_tienda.upper()}",
              fontsize=14, fontweight='bold', pad=25, color='#2C3E50', loc='left')

  for barra in barras:
    ancho = barra.get_width()
    plt.text(ancho + (ancho * 0.01),
             barra.get_y() + barra.get_height()/2,
             f'{int(ancho)} uds',
             va='center', fontsize=11, fontweight='bold', color='#2C3E50')
  ax = plt.gca()
  ax.invert_yaxis()
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.spines['bottom'].set_color('#CCCCCC')
  ax.spines['left'].set_color('#CCCCCC')

  plt.tight_layout()
  plt.show()



```

#**5. Envío promedio por tienda**

```python
costo_envio_promedio_por_tienda = {
    "tienda 1": tienda["Costo de envío"].mean().round(2),
    "tienda 2": tienda2["Costo de envío"].mean().round(2),
    "tienda 3": tienda3["Costo de envío"].mean().round(2),
    "tienda 4": tienda4["Costo de envío"].mean().round(2)
}
# Ordenar el diccionario para que la respuesta se ordene de menor a mayor

promedio_en_orden = sorted(costo_envio_promedio_por_tienda.items(), key=lambda item: item[1], reverse=False)

for numero_de_la_tienda, promedio_de_envio in promedio_en_orden:
  print(f'El gasto promedio de servicio de envío de la {numero_de_la_tienda} es de {promedio_de_envio:,.2f} ')

```
