```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

```

# **1. Análisis de facturación**

```python
total_ganancia_por_tienda = {
    "tienda 1": tienda["Precio"].sum(),
    "tienda 2": tienda2["Precio"].sum(),
    "tienda 3": tienda3["Precio"].sum(),
    "tienda 4": tienda4["Precio"].sum()
}

for numero_de_tienda, total in total_ganancia_por_tienda.items(): #.item dice: oye no solo me des la llave pasame todo lo que hay ahí, o sea la operación matematica
  print(f'El ingreso total de la {numero_de_tienda} es de: ${total:,.2f}') #,.2f le ponemos comas a los miles y solo 2 decimales
```

# **2. Ventas por categoría**







     




```python
# tienda.isnull().sum() comprobamos que no hay datos nulos.

cantidad_productos_vendidos = {
   "tienda 1": tienda['Categoría del Producto'].value_counts().head(3).reset_index(),
   "tienda 2":tienda2['Categoría del Producto'].value_counts().head(3).reset_index(),
   "tienda 3":tienda3['Categoría del Producto'].value_counts().head(3).reset_index(),
   "tienda 4":tienda4['Categoría del Producto'].value_counts().head(3).reset_index()
}

for nombre_tienda, cantidad_productos in cantidad_productos_vendidos.items():
  print(f'\n---Las 3 categorías más vendidas en la {nombre_tienda} son:---')
  print(cantidad_productos)
```

#**3. Calificación promedio de la tienda**

```python
promedio_por_tienda = {
    "tienda 1": tienda["Calificación"].mean().round(2),
    "tienda 2": tienda2["Calificación"].mean().round(2),
    "tienda 3": tienda3["Calificación"].mean().round(2),
    "tienda 4": tienda4["Calificación"].mean().round(2)
}
# Ordenar el diccionario para que la respuesta se ordene de mayor a menor

promedio_ordenado = sorted(promedio_por_tienda.items(), key=lambda item: item[1], reverse=True)

for nombre_de_tienda, promedio_de_tienda in promedio_ordenado:
  print(f'La satisfacción del cliente en relación con los productos vendidos de la {nombre_de_tienda} es de {promedio_de_tienda} ')
```

#**4. Productos más y menos vendidos**

```python
maximo_y_minimo = {
    "tienda 1": pd.concat([tienda['Producto'].value_counts().head(3), tienda['Producto'].value_counts().tail(3)]).reset_index(),
    "tienda 2": pd.concat([tienda2['Producto'].value_counts().head(3), tienda2['Producto'].value_counts().tail(3)]).reset_index(),
    "tienda 3": pd.concat([tienda3['Producto'].value_counts().head(3), tienda3['Producto'].value_counts().tail(3)]).reset_index(),
    "tienda 4": pd.concat([tienda3['Producto'].value_counts().head(3), tienda3['Producto'].value_counts().tail(3)]).reset_index()
}

for titulo_tienda, tabla in maximo_y_minimo.items():
  print(f'\n --- Productos más y menos vendidos de la {titulo_tienda} son:  ')
  print(f'{tabla}')

#tienda['Producto'].value_counts().head(3).tail(3).reset_index()
```

#**5. Envío promedio por tienda**

```python
costo_envio_promedio_por_tienda = {
    "tienda 1": tienda["Costo de envío"].mean().round(2),
    "tienda 2": tienda2["Costo de envío"].mean().round(2),
    "tienda 3": tienda3["Costo de envío"].mean().round(2),
    "tienda 4": tienda4["Costo de envío"].mean().round(2)
}
# Ordenar el diccionario para que la respuesta se ordene de mayor a menor

promedio_en_orden = sorted(costo_envio_promedio_por_tienda.items(), key=lambda item: item[1], reverse=False)

for numero_de_la_tienda, promedio_de_envio in promedio_en_orden:
  print(f'El gasto promedio de servicio de envío de la {numero_de_la_tienda} es de {promedio_de_envio:,.2f} ')
```
