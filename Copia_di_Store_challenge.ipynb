```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda4.shape
```

#**Ingreso total por tienda**

```python
total_ganancia_por_tienda = {
    "tienda 1": tienda["Precio"].sum(),
    "tienda 2": tienda2["Precio"].sum(),
    "tienda 3": tienda3["Precio"].sum(),
    "tienda 4": tienda4["Precio"].sum()
}

for numero_de_tienda, total in total_ganancia_por_tienda.items(): #.item dice: oye no solo me des la llave pasame todo lo que hay ahí, o sea la operación matematica
  print(f'El ingreso total de la {numero_de_tienda} es de: ${total:,.2f}') #,.2f significa: Ponle comas para los miles y solo 2 decimales
```

# **1. Análisis de facturación**



     




#**2. Ventas por categoría**

```python
# tienda.isnull().sum() comprobamos que no hay datos nulos.

cantidad_productos_vendidos = {
   "tienda 1": tienda['Categoría del Producto'].value_counts().head(3).reset_index(),
   "tienda 2":tienda2['Categoría del Producto'].value_counts().head(3).reset_index(),
   "tienda 3":tienda3['Categoría del Producto'].value_counts().head(3).reset_index(),
   "tienda 4":tienda4['Categoría del Producto'].value_counts().head(3).reset_index()
}

for nombre_tienda, cantidad_productos in cantidad_productos_vendidos.items():
  print(f'\n---Los 3 productos más vendidos en la {nombre_tienda} son:---')
  print(cantidad_productos)
```

#**3. Calificación promedio de la tienda**

```python
promedio_calificacion = {
    "tienda 1": tienda.groupby("Categoría del Producto").agg({"Calificación": ['mean']}).round(2),
    "tienda 2": tienda2.groupby("Categoría del Producto").agg({"Calificación": ['mean']}).round(2),
    "tienda 3": tienda3.groupby("Categoría del Producto").agg({"Calificación": ['mean']}).round(2),
    "tienda 4": tienda4.groupby("Categoría del Producto").agg({"Calificación": ['mean']}).round(2)
}

for nombre_de_tienda, promedio_por_tienda in promedio_calificacion.items():
  print(f'\n --- La satisfacción del cliente en relación con los productos vendidos de la {nombre_de_tienda}: ')
  print(f'{promedio_por_tienda}')
```
